# Visualizing Earnings Based On College Majors

## Headers for  recent-grads.csv

Header | Description
---|---------
`Rank` | Rank by median earnings
`Major_code` | Major code, FO1DP in ACS PUMS
`Major` | Major description
`Major_category` | Category of major from Carnevale et al
`Total` | Total number of people with major
`Sample_size` | Sample size (unweighted) of full-time, year-round ONLY (used for earnings)
`Men` | Male graduates
`Women` | Female graduates
`ShareWomen` | Women as share of total
`Employed` | Number employed (ESR == 1 or 2)
`Full_time` | Employed 35 hours or more
`Part_time` | Employed less than 35 hours
`Full_time_year_round` | Employed at least 50 weeks (WKW == 1) and at least 35 hours (WKHP >= 35)
`Unemployed` | Number unemployed (ESR == 3)
`Unemployment_rate` | Unemployed / (Unemployed + Employed)
`Median` | Median earnings of full-time, year-round workers
`P25th` | 25th percentile of earnigns
`P75th` | 75th percentile of earnings
`College_jobs` | Number with job requiring a college degree
`Non_college_jobs` | Number with job not requiring a college degree
`Low_wage_jobs` | Number in low-wage service jobs

```python
from IPython.core.interactiveshell import InteractiveShell
from IPython.display import Markdown
import pandas
import matplotlib
import matplotlib.pyplot as plot
%matplotlib inline
InteractiveShell.ast_node_interactivity = "all"

recent_grads = pandas.read_csv(open("data/recent-grads.csv", "r"))
recent_grads = recent_grads.dropna()
Markdown("## Data overview")
Markdown("### Head")
recent_grads.head(2)
Markdown("### Tail")
recent_grads.tail(2)
Markdown("### Describe")
recent_grads.describe()
```

```python
recent_grads.plot(kind='scatter', x='Sample_size', y='Median', title='Sample size vs Median')
```

```python
recent_grads.plot(kind='scatter', x='Sample_size', y='Unemployment_rate', title='Sample size vs Unemployment rate')
```

```python
recent_grads.plot(kind='scatter', x='Full_time', y='Median', title='Full-time vs Median')
```

```python
recent_grads.plot(kind='scatter', x='ShareWomen', y='Unemployment_rate', title='Share Women vs Unemployment ate')
```

```python
recent_grads.plot(kind='scatter', x='Men', y='Median', title='Men vs Median')
```

```python
recent_grads.plot(kind='scatter', x='Women', y='Median', title='Women vs Median')
```

```python
cols = ["Sample_size", "Median", "Employed", "Full_time", "ShareWomen", "Unemployment_rate", "Men", "Women"]

fig = plot.figure(figsize=(5,12))
for r in range(1,5):
    ax = fig.add_subplot(4,1,r)
    ax = recent_grads[cols[r]].plot(kind='hist', rot=40)
```

### Questions

#### Do students in more popular majors make more money?

```python
recent_grads.plot(kind='scatter', x='Total', y='Median', title='Total vs Median')
```

It seems that as more popular the major is, less money the person makes.

#### Do students that majored in subjects that were majority female make more money?

```python
recent_grads.plot(kind='scatter', x='ShareWomen', y='Median', title='Share Women vs Median')
```

It seems that majored people in subjects where the majority of women are lower make more money.

#### Is there any link between the number of full-time employees and median salary?

```python
recent_grads.plot(kind='scatter', x='Full_time', y='Median', title='Full-time vs Median')
```

It doesn't seem that people in majors with more full-time workers make more money.
